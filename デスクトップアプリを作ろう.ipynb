# ウィンドウに画像を表示

ウィンドウに画像を表示する流れは、次の通りです。

キャンバスを作る→画像を作る→キャンバス上に画像を配置する

キャンパスを作るには、tkinterの「Canvas」を使います。

Canvas
- bg：背景を設定
- width：キャンパスの横幅
- height：キャンパスの縦幅

キャンパスの置く位置を指定するには、Canvasオブジェクトのplaceメソッドを使います。

画像を作るには、tkinterのPhotoImageを使います。
- file：ファイルパス

キャンバスに画像を配置するには、Canvasオブジェクトのcreate_imageメソッドを使います。
- 1つ目の引数：x座標
- 2つ目の引数：y座標
- image：画像データ

```python
import tkinter

# ウィンドウ作成
window = tkinter.Tk()
window.title("AIボット")
window.minsize(470, 630)

# 画像表示
canvas = tkinter.Canvas(bg="white", width=470, height=630)
canvas.place(x=0, y=0)

# フレーム画像の設定
frame_img = tkinter.PhotoImage(file ="./images/tablet.png")
canvas.create_image(235, 315, image = frame_img)

# チャットボットイメージ画像の設定
ai_img = tkinter.PhotoImage(file= "./images/ai-normal.png")
canvas.create_image(235, 280, image=ai_img, tag="ai")

# メインループ
window.mainloop()
```

ウィンドウに文字を表示するにはtkinterの「Label」、入力欄を作るには「Entry」、ボタンを作るには「Button」を使います。

Label
- text：表示する文字を指定
- bg：背景を設定

入力欄を作るには、tkinterのEntryを使います。

Entry
- width：横幅
- bd：テキストボックスの線の幅

Button

文字の見た目を調整するには、Tkオブジェクトのoption_addを使います。

option_addはウィンドウに設定を追加するときに使うメソッドです。

# AIチャットボットフレーム

```python
import tkinter
import tkinter.font as font
from time import sleep
# ウィンドウ作成
window = tkinter.Tk()
window.title("AIボット")
window.minsize(470, 630)
# フォントの設定
window.option_add("*font", ["MSゴシック", 24])
# 画像表示
canvas = tkinter.Canvas(bg="white", width=470, height=630)
canvas.place(x=0, y=0)
# フレーム画像の設定
frame_img = tkinter.PhotoImage(file="./images/tablet.png")
canvas.create_image(235, 315, image=frame_img)
# チャットボットイメージ画像の設定
img_list = []
img_list.append(tkinter.PhotoImage(file="./images/ai-normal.png"))
img_list.append(tkinter.PhotoImage(file="./images/ai-happy.png"))
canvas.create_image(235, 280, image=img_list[0], tag="ai-bot")
#ボタン
askbutton = tkinter.Button(text="▲", font=("", 18))
askbutton.place(x=386, y=511)
#出力エリア
canvas.create_rectangle(40, 420, 430, 500, outline = "pink", fill = "white")
answer_font = font.Font(window,family="HG創英角ポップ体",size=16)
answer = tkinter.Label(text="(話しかけられるのを待ってる)", bg="white", font=answer_font)
answer.place(x=50, y=430)
#入力エリア
inputarea = tkinter.Entry(width=21, bd=4)
inputarea.place(x=34, y=511)
#エンターキー押した
def ask_enter(self):
  ask_click()
#ボタン押した
def ask_click():
  answer["text"] = "connecting"
  for num in range(3):
    window.update()
    answer["text"] += "."

    sleep(0.5)
# 入力値の取得
message = inputarea.get()
# 入力欄クリア
inputarea.delete(0, tkinter.END)
# TODO ルールベースでメッセージのやり取りを行う。
if message == "おはよう":
  canvas.itemconfig("ai-bot", image=img_list[1])
  answer["text"] = "おはよう!!"
else:
  canvas.itemconfig("ai-bot", image=img_list[0])
  answer["text"] = "(話しかけられるのを待ってる)"
#ボタンに紐づけ
askbutton["command"] = ask_click
#リターンキー対応
inputarea.bind('<Return>', ask_enter)
# メインループ
window.mainloop()
```
